## Dataset Overview

This dataset provides detailed information on the environmental and logistical aspects of product supply across various locations, consisting of 10,000 entries. It is designed for in-depth analysis of carbon emissions, sustainability, and supply chain efficiency. The dataset includes the following columns:

1. **Supplier Name**: The name of the supplier providing the product.
2. **Location**: The geographical location of the supplier or distribution center.
3. **Category**: The category of the product.
4. **Product**: The specific product being analyzed.
5. **Transportation Mode**: The method of transportation used (e.g., Air, Train, Ship, Truck).
6. **Transportation Route**: The route taken for transportation.
7. **Distribution Network**: The network through which the product is distributed.
8. **Emission Intensity (kg CO2/unit)**: The amount of CO2 emissions produced per unit of the product.
9. **Total Emissions (ton CO2)**: The total CO2 emissions associated with the product.
10. **Fuel Consumption (liters)**: The amount of fuel consumed during transportation.
11. **Energy Consumption (GJ)**: The total energy consumed during the process.
12. **Sustainability Rating**: A rating indicating the environmental sustainability of the product.
13. **Regulatory Impact**: The impact of regulatory requirements on the product.
14. **Year**: The year in which the data was recorded.

### Purpose

This dataset is intended for analyzing carbon emissions and sustainability in supply chains, allowing users to:

- Identify high-impact areas for carbon footprint reduction.
- Evaluate the efficiency of different transportation modes and routes.
- Assess the overall sustainability of products and their supply chains.
- Develop strategies to minimize environmental impact and improve operational efficiency.

**<h3>Transport Mode:</h3>**

```python
# Calculate the emission intesity for each row
df['Calculated Emission Intensity'] = df['Total Emissions (ton CO2)'] * 1000 / df['Fuel Consumption (liters)']

# Aggregate by Tanportation Mode
mode_intensity = df.groupby('Transportation Mode').agg({
    'Emission Intensity (kg CO2/unit)':'mean',
    'Total Emissions (ton CO2)':'sum',
    'Fuel Consumption (liters)':'sum',
    'Calculated Emission Intensity':'mean'
}).reset_index()

mode_intensity
```

```python
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

# Bar Chart 1: Total Emissions by Transportation Mode
axs[0, 0].bar(mode_intensity['Transportation Mode'], mode_intensity['Total Emissions (ton CO2)'], color='darkgoldenrod')
axs[0, 0].set_xlabel('Transportation Mode')
axs[0, 0].set_ylabel('Total Emissions (ton CO2)')
axs[0, 0].set_title('Total Emissions by Transportation Mode')

# Bar Chart 2: Calculated Emission Intensity by Transportation Mode
axs[0, 1].bar(mode_intensity['Transportation Mode'], mode_intensity['Calculated Emission Intensity'], color='darkslategray')
axs[0, 1].set_xlabel('Transportation Mode')
axs[0, 1].set_ylabel('Calculated Emission Intensity (kg CO2/unit)')
axs[0, 1].set_title('Calculated Emission Intensity by Transportation Mode')

# Line Chart 1: Emission Intensity by Transportation Mode
axs[1, 0].plot(mode_intensity['Transportation Mode'], mode_intensity['Emission Intensity (kg CO2/unit)'], marker='o', linestyle='-', color='purple')
axs[1, 0].set_xlabel('Transportation Mode')
axs[1, 0].set_ylabel('Emission Intensity (kg CO2/unit)')
axs[1, 0].set_title('Emission Intensity by Transportation Mode')
axs[1, 0].grid(True)

# Line Chart 2: Fuel Consumption by Transportation Mode
axs[1, 1].plot(mode_intensity['Transportation Mode'], mode_intensity['Fuel Consumption (liters)'], marker='o', linestyle='-', color='crimson')
axs[1, 1].set_xlabel('Transportation Mode')
axs[1, 1].set_ylabel('Fuel Consumption (liters)')
axs[1, 1].set_title('Fuel Consumption by Transportation Mode')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()

# Save the chart
plt.savefig('Emission Intensity of Transport Mode.png')

# Show the plots
plt.show()
```

**<h3>Supplier:</h3>**

```python
# Find out Sustainability Rating of each supplier
sustainability_rating = df.groupby('Supplier Name').agg({'Total Emissions (ton CO2)':'sum','Sustainability Rating':'mean'}).reset_index().round(2)
sustainability_rating.head()
```

```python
# Identify top 10 suppliets for both Total Emission and Sustainability Rating
top10_emissions = sustainability_rating.nlargest(10, 'Total Emissions (ton CO2)')['Supplier Name']
top10_rating = sustainability_rating.nlargest(10,'Sustainability Rating')['Supplier Name']

# Create a figure and a set of subplots
fig,ax1 = plt.subplots(figsize=(12,8))

# Bar chart for Total Emission
bars = ax1.bar(sustainability_rating['Supplier Name'],sustainability_rating['Total Emissions (ton CO2)'], color='skyblue',label='Total Emissions (ton CO2)')

# Highlight top 10 bar
for bar, supplier in zip(bars, sustainability_rating['Supplier Name']):
    if supplier in top10_emissions.values:
        bar.set_color('blue') # Highlight Color

ax1.set_xlabel('Supplier Name')
ax1.set_ylabel('Total Emission (ton CO2)', color='black')
ax1.tick_params(axis='x',rotation=90)
ax1.tick_params(axis='y',labelcolor='black')

# Create a second Y Axis for line chart
ax2 = ax1.twinx()
line, = ax2.plot(sustainability_rating['Supplier Name'],sustainability_rating['Sustainability Rating'], color='green', marker='o',label='Sustainability Rating')

# Highligh top 10 in the line chart
for x, y, supplier in zip(sustainability_rating['Supplier Name'], sustainability_rating['Sustainability Rating'], sustainability_rating['Supplier Name']):
    if supplier in top10_rating.values:
        ax2.plot(x,y, color='red', marker='o',markersize=10) # Highlight top 10 with larger red markers

ax2.set_ylabel('Sustainability Rating', color='black')
ax2.tick_params(axis='y',labelcolor='black')

# Title and layout adjustments
plt.title('Total Emissions vs Sustainability Rating by Supplier with Top 10 Highlighted')
fig.tight_layout()

#Save the layout
plt.savefig('Total Emissions vs Sustainability Rating by Supplier with Top 10 Highlighted.png')

# Show the plot
plt.show()
```

**<h3>Distribution Routes:</h3>**

```python
# Find out the most carbon intensive distribution network
emission_by_distribution = df.groupby('Distribution Network')['Emission Intensity (kg CO2/unit)'].sum()
emission_by_distribution = emission_by_distribution.sort_values(ascending=False)
emission_by_distribution
```

The distribution network shows significant variation in emission intensity. Suppliers have the highest emissions at 18,605.36 kg CO2/unit, while Retailers, at the lower end, contribute 16,677.76 kg CO2/unit. Notably, Logistics and Wholesalers are close behind Suppliers, whereas Manufacturers and Warehouses have slightly lower emissions, just above Retailers.

```python
emission_by_routes = df.groupby('Transportation Route').agg({
    'Emission Intensity (kg CO2/unit)': 'sum',
    'Fuel Consumption (liters)': 'sum'
}).reset_index()

# Sort by 'Fuel Consumption (liters)' in descending order and then by 'Emission Intensity (kg CO2/unit)' in ascending order
emission_by_routes = emission_by_routes.sort_values(by=['Fuel Consumption (liters)', 'Emission Intensity (kg CO2/unit)'], ascending=[False, True])

top10_routes = emission_by_routes.head(10)
top10_routes
```

```python
bottom10_routes = emission_by_routes.tail(10)
bottom10_routes
```

```python
# Creaate bar plot
fig, ax1 = plt.subplots(figsize=(10,8))

# Plot bars with different color
ax1.bar(top10_routes['Transportation Route'],top10_routes['Emission Intensity (kg CO2/unit)'],color='gray', label='Top 10 Emission Intensity')
ax1.bar(bottom10_routes['Transportation Route'], bottom10_routes['Emission Intensity (kg CO2/unit)'], color='seagreen', label='Bottom 10 Emission Intensity')
ax1.set_xlabel('Transportatin Route')
ax1.set_ylabel('Emission Intensity (kg CO2/unit)')
ax1.set_title('Emission Intensity and Fuel Consumption by Transportation Route')
ax1.tick_params(axis='x', rotation=90)


# Create a second axis
ax2 = ax1.twinx()

# Combine both top and bottom for line plot
combined_routes = pd.concat([top10_routes,bottom10_routes])

# Line plot for fuel consumption
sns.lineplot(x=combined_routes['Transportation Route'], y=combined_routes['Fuel Consumption (liters)'], marker='o', color='blue', ax=ax2, label='Fuel Consumption')
ax2.set_ylabel('Fuel Consumption (liters)')

# Add legend
ax1.legend(loc='upper right')
ax2.legend(loc='upper center')

plt.tight_layout()

# Save the image 
plt.savefig('Emission Intensity and Fuel Consumption by Transportation Route.png')


# Show the plot
plt.show()
```

**<h3>Regulatory Impacts:</h3>**

```python
# Analyze potential reduction if regulatory compliance is improved
# For simplicity, assume that improvement in regulatory compliance means reducing emissions by a fixed percentage (e.g., 12%)
reduction_percentage = 0.12  # 12% reduction

# Calculate potential reduction for each regulatory impact category
df['Potential Reduction (ton CO2)'] = df['Total Emissions (ton CO2)'] * reduction_percentage

# Calculate the average potential reduction for each regulatory impact category
average_reduction = df.groupby('Regulatory Impact')['Potential Reduction (ton CO2)'].mean()

# Display the average potential reduction by regulatory impact
print("\nAverage Potential Reduction by Regulatory Impact:")
print(average_reduction)
```

```python
# Review which regulations have the most significant impact on emissions
regulatory_compliance_summary = df.groupby('Regulatory Impact').agg({
    'Total Emissions (ton CO2)': 'sum',
    'Emission Intensity (kg CO2/unit)': 'mean',
    'Potential Reduction (ton CO2)':'mean'
}).reset_index().round(2)

regulatory_compliance_summary
```

```python
# Plot
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar chart for Total Emissions
color = 'tab:blue'
ax1.bar(regulatory_compliance_summary['Regulatory Impact'], 
        regulatory_compliance_summary['Total Emissions (ton CO2)'],
       color=color, alpha=0.6, label='Total Emissions (ton CO2)')
ax1.set_xlabel('Regulatory Impact')
ax1.set_ylabel('Total Emissions (ton CO2)', color=color)
ax1.tick_params(axis='y',labelcolor=color)
ax1.set_title('Total Emissions, Emission Intensity, and Potential Reduction by Regulatory Impact')

# Create a second y axis for Emission Intensity
ax2 = ax1.twinx()
color = 'tab:green'
ax2.plot(regulatory_compliance_summary['Regulatory Impact'], 
         regulatory_compliance_summary['Emission Intensity (kg CO2/unit)'], 
         color=color, marker='o', label='Emission Intensity (kg CO2/unit)')
ax2.set_ylabel('Emission Intensity (kg CO2/unit)', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create a third y-axis for Poteintial Reductionabs
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward',60)) # Offset the third axis
color = 'tab:red'
ax3.plot(regulatory_compliance_summary['Regulatory Impact'], 
         regulatory_compliance_summary['Potential Reduction (ton CO2)'], 
         color=color, marker='o', linestyle='-', label='Potential Reduction (ton CO2)')
ax3.set_ylabel('Potential Reduction (ton CO2)', color=color)
ax3.tick_params(axis='y', labelcolor=color)

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax3.legend(loc='upper center')
plt.tight_layout()

# Save the plot 
plt.savefig('Total Emissions, Emission Intensity, and Potential Reduction by Regulatory Impact.')


# Show the plot
plt.show()
```

### Comparative Narrative:

The analysis of regulatory impact on emissions reveals that the **High Regulatory Impact** category leads in total emissions with **85,662.29 tons of CO2**, followed by the **Low** and **Medium** categories with **83,952.91 tons** and **81,748.23 tons of CO2**, respectively. Despite the High Regulatory Impact category having the highest total emissions, its emission intensity is the lowest at **10.39 kg CO2 per unit**, indicating effective per-unit emission control measures. Conversely, the **Medium Regulatory Impact** category, although having the lowest total emissions, exhibits the highest emission intensity at **10.60 kg CO2 per unit**, suggesting inefficiencies in emission control on a per-unit basis.

Regarding potential CO2 reduction, the figures are relatively consistent across all categories, with reductions ranging from **3.00 to 3.03 tons of CO2**. This consistency indicates that, irrespective of the regulatory impact level, similar reductions can be achieved with targeted interventions.

Strengthening regulations in these areas, such as by imposing stricter emission limits or incentivizing cleaner production methods, could help reduce overall emissions more effectively.

- **Maintain and Improve High Impact Regulations**:
